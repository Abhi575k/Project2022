```python
from matplotlib import pyplot as plt
import numpy as np
```

### pie chart


```python
y = np.array([26, 19, 14, 7])
labels = ["five star", "dairy milk", "kit kat", "nesley"]

plt.pie(y, labels = labels)
plt.title("top categories of chocolates by samith")
plt.show()
```

### line graph


```python
x1 = np.array(['day 1', 'day 2', 'day 3', 'day 4'])
y1 = np.array([5, 6, 7, 8])
y2 = np.array([9, 2, 6, 9])
y3 = np.array([2, 5, 3, 7])

plt.plot(x1,y1, marker='x')
plt.plot(x1,y2, marker='s')
plt.plot(x1,y3, marker='>')
plt.title("number of chocolates per student")
plt.show()
```

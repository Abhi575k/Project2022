### Matplotlib
#### Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### line graph

```python
x1 = np.array(['day 1', 'day 2', 'day 3', 'day 4'])
y1 = np.array([7, 6, 9, 8])
y2 = np.array([8, 2, 6, 9])
y3 = np.array([2, 5, 3, 7])

plt.plot(x1,y1, marker='x')
plt.plot(x1,y2, marker='s')
plt.plot(x1,y3, marker='>')
plt.title("cups of coffee per employee")
plt.show()
```

#### Bar graph

```python
x2 = np.array(['day 1','day 2','day 3','day 4'])
y2 = np.array([17, 13, 18, 24])
plt.bar(x2, y2, width=0.6, color="red")
plt.title("Total cups of coffee per day" )
plt.show()
```

#### Pie chart

```python
x=np.array([8,2,6,9])
plt.pie(x, labels=['shaun','mark','david','john doe'])
plt.title("wednesday coffee consumption")
plt.show()
```

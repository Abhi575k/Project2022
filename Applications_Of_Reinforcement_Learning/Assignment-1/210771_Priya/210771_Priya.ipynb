```python
import numpy as np

import time
start = time.process_time()
   

##ADDITION
P = int(input("Enter the number of rows:"))
Q = int(input("Enter the number of columns:"))

A=[]

for i in range(P):      
    a =[]
    for j in range(Q):      
         a.append(int(input()))
    A.append(a)

B=[]
for i in range(P):
    b=[]
    for j in range(Q):
        b.append(int(input()))
    B.append(b)    

add=[]
for i in range(P):
    c=[]
    for j in range(Q):
        c.append(0)
    add.append(c)
    
    
for i in range(0,len(A)):
    for j in range(0,len(B[0])):
        add[i][j]= A[i][j] + B[i][j]
for x in add:
   print(x)
print(time.process_time() - start)

print(np.add(A,B))
```

```python


#MULTIPLICATION
start = time.process_time()

m = int(input("Enter the number of rows:"))
n = int(input("Enter the number of columns1:"))
r = int(input("Enter the number of columns2:"))

C=[]

for i in range(m):      
    a =[]
    for j in range(n):      
         a.append(int(input()))
    C.append(a)

D=[]
for i in range(n):
    b=[]
    for j in range(r):
        b.append(int(input()))
    D.append(b)    

multi=[]
for i in range(m):
    c=[]
    for j in range(r):
        c.append(0)
    multi.append(c)

for i in range(len(C)):
    for j in range(len(D[0])):
        for k in range(len(C[0])):
            multi[i][j]+= C[i][k]*D[k][j]
for z in multi:
        print(z)
print(time.process_time() - start)

print( np.dot(C,D))
```

# multiplication of matrix

```python
import numpy as np
import time
s=time.time()
x=int(input("enter rows:"))
y=int(input("enter cols:"))
y1=int(input("entre col for m2:"))
m1=[]
for i in range(x):
    a=[]
    for j in range(y):
        a.append(int(input("for m1:")))
    m1.append(a)
m2=[]
for i in range(y):
    c=[]
    for j in range(y1):
        c.append(int(input("for m2:" )))
    m2.append(c)
    
ans=[]

for i in range(x):
    r=[]
    for j in range(y1):
        r.append(0)
    ans.append(r)
        
        
for i in range(len(m1)) :
    for j in range(len(m2[0])):
        for k in range(len(m1[0])):
            ans[i][j]+=m1[i][k]*m2[k][j]
            
            
for r in ans:
    print(r)
print("\n")  
e=time.time()
print(e-s)
```

```python

print(np.dot(m1,m2))
```

# addition of matrix

```python
import time
s=time.time()
n=int(input("no of rows:"))
m=int(input("no. of cols:"))
m2=[]
for i in range(n):
    c=[]
    for j in range(m):
        c.append(int(input("for m2:")))
    m2.append(c)
m1=[]
for i in range(n):
        c=[]
        for j in range(m):
            c.append(int(input("for m1:")))
        
        m1.append(c)

ans=[]
for i in range(n):
        c=[]
        for j in range(m):
            c.append(0)
        ans.append(c)
        
for i in range(n):
    for j in range(m):
        ans[i][j]=m1[i][j]+m2[i][j]
        
for r in ans:
    print(r)
e=time.time()
print(e-s)

```

```python
print(np.add(m1,m2))
```

# matrix inverse

```python
def trans(m):
    return map(list,zip(*map))
def minor(m,i,j):
    return [row[:j]+row[j+1:] 
           for row in (m[:i]+m[i+1:])]
def det(m):
    if len(m)==2:
        return m[0][0]*m[1][1]-m[0][1]*m[1][0]
    det=0
    for c in range(len(m)):
        det+=((-1)**c)*m[0][c]*det(minor(m,0,c))
        return det
def inverse(m):
    dete=det(m)
    if len(m)==2:
        return [m[1][1]/dete,-1*m[0][1]/dete,-1*m[1][0]/dete,m[0][0]/dete]
    
    cfs=[]
    for r in range(len(m)):
        cfr=[]
        for c in range(len(m)):
            minor=minor(m,r,c)
            cfr.append((-1)**(r+c))*det(minor)
        cfs.append(cfr)
    cfs=trans(cfs)
    for r in range(len(cfs)):
        for c in range(len(cfs)):
            cfs[r][c]=cfs[r][c]/dete
    return cfs

m=[[4,3],[8,5]]
print(inverse(m))
```

```python
print(np.linalg.inv(m))
```

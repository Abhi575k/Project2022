```python
from numpy import random
import numpy as np
```

```python
m = random.randint(100)
n = random.randint(100)
p = random.randint(100)
x = random.randint(100, size=(m,n))
y = random.randint(100, size=(n,p))
```

```python
#Without using numpy
mat_multiply = random.randint(1, size=(m,p))
for i in range(m):
   for j in range(p):
      for k in range(n):
        mat_multiply[i][j] += x[i][k]*y[k][j]

print(mat_multiply)
```

```python
#Using numpy
ans = np.dot(x,y)
print(ans)
```

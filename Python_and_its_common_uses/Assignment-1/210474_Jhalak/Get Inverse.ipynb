```python
#inverse of matrix
import numpy as np
def getCofactor(mat,temp,k,l,n):
  i = 0 
  j = 0
  row = 0
  col = 0
  for row in range(len(mat)):
      for col in range(len(mat[0])):
       if row != k and col != l:
           temp[i][j] = mat[row][col]
           j= j+1
           if j == len(mat)-1:
              j = 0
              i = i+1
def determinant(mat,n):
    Det = 0
    if n == 1:
      return mat[0][0]
    else:
        temp = [[0 for x in range(n)] for y in range(n)]
        sign = 1
        f = 0
        for f in range(n):
            getCofactor(mat,temp,0,f,n)
            Det += sign*mat[0][f]*determinant(temp,n-1)
            sign = - sign

    return Det
def adjoint_fn(mat,adjoint,n):
    if n == 1:
        adjoint[0][0] = 1
    sign = 1
    temp = [[0 for x in range(n)] for y in range(n)]  
    i = 0
    j = 0
    for i in range(n):
        for  j in range(n):
            getCofactor(mat,temp,i,j,n)
            if (i+j)%2 == 0:
             sign = 1
            else: 
             sign = -1
            
            adjoint[j][i] = (sign)*(determinant(temp,n-1))

def inverse_fn(mat,inverse,n):
    det = determinant(mat,n)

    if det == 0:
        print("Inverse doesn't exist")

    else:
        adjoint = [[0 for x in range(n)] for y in range(n)]
        adjoint_fn(mat,adjoint,n)
        i=0
        j=0
        for i in range(n):
            for j in range(n):
                inverse[i][j] = adjoint[i][j]/det

n = int(input("Order of matrix: "))
mat = np.random.randint(100,size=(n,n))
adjoint = [[0 for x in range(n)] for y in range(n)]
inverse = [[0 for x in range(n)] for y in range(n)]
inverse_fn(mat,inverse,n)
print(mat) 
print(inverse)

```
